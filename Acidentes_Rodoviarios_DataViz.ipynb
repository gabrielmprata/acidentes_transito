#**<font color=#4c60d6 size="6"> 3. Construção dos gráficos para o anuário de 2024 📊**

###**<font color=#4c60d6> 3.1 Sinistros de Trânsito no Brasil**

####**<font color=#4c60d6> 3.1.1 Quadro comparativo, 2024-2023**

```python
# agrupa por ano
df_ind_anual = df_hs_acidentes.groupby('ano')[['sinistro','pessoas','mortos', 'feridos_leves', 'feridos_graves', 'ilesos', 'ignorados', 'veiculos']].apply(lambda x: x.sum()).reset_index()
```

```python
sinistro_atual = df_ind_anual.sinistro.values[1]
sinistro_ant = df_ind_anual.sinistro.values[0]
sinistro_delta = sinistro_atual - sinistro_ant

veiculos_atual = df_ind_anual.veiculos.values[1]
veiculos_ant = df_ind_anual.veiculos.values[0]
veiculos_delta = veiculos_atual - veiculos_ant

pessoas_atual = df_ind_anual.pessoas.values[1]
pessoas_ant = df_ind_anual.pessoas.values[0]
pessoas_delta = pessoas_atual - pessoas_ant

feridos_leves_atual = df_ind_anual.feridos_leves.values[1]
feridos_leves_ant = df_ind_anual.feridos_leves.values[0]
feridos_leves_delta = feridos_leves_atual - feridos_leves_ant

feridos_graves_atual = df_ind_anual.feridos_graves.values[1]
feridos_graves_ant = df_ind_anual.feridos_graves.values[0]
feridos_graves_delta = feridos_graves_atual - feridos_graves_ant

mortos_atual = df_ind_anual.mortos.values[1]
mortos_ant = df_ind_anual.mortos.values[0]
mortos_delta = mortos_atual - mortos_ant

ilesos_atual = df_ind_anual.ilesos.values[1]
ilesos_ant = df_ind_anual.ilesos.values[0]
ilesos_delta = ilesos_atual - ilesos_ant
```

```python
fig = go.Figure()

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = float(df_ind_anual.sinistro.values[1]),
    delta = {"reference": int(df_ind_anual.sinistro.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Sinistros" },
    domain = {'row': 0, 'column': 0}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.veiculos.values[1]),
    delta = {"reference": int(df_ind_anual.veiculos.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Veículos" },
    domain = {'row': 0, 'column': 1}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.pessoas.values[1]),
    delta = {"reference": int(df_ind_anual.pessoas.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Pessoas" },
    domain = {'row': 1, 'column': 0}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.ilesos.values[1]),
    delta = {"reference": int(df_ind_anual.ilesos.values[0]),'increasing': {'color': "green"},'decreasing': {'color': "red"}},
    title = {"text": "Ilesos" },
    domain = {'row': 1, 'column': 1}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.feridos_leves.values[1]),
    delta = {"reference": int(df_ind_anual.feridos_leves.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Feridos Leves" },
    domain = {'row': 2, 'column': 0}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.feridos_graves.values[1]),
    delta = {"reference": int(df_ind_anual.feridos_graves.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Feridos Graves" },
    domain = {'row': 2, 'column': 1}))

fig.add_trace(go.Indicator(
    mode = "number+delta",
    value = int(df_ind_anual.mortos.values[1]),
    delta = {"reference": int(df_ind_anual.mortos.values[0]),'increasing': {'color': "red"},'decreasing': {'color': "green"}},
    title = {"text": "Óbitos" },
    domain = {'row': 3, 'column': 0}))

fig.update_layout(
    paper_bgcolor = "lightgray",
    width=650, height = 800,
    grid = {'rows': 4, 'columns': 2, 'pattern': "independent"},
    )
```

####**<font color=#4c60d6> 3.1.2 Série histórica, 2007-2024**

```python
df_hs_anual = pd.read_csv(url3, delimiter=';')
df_hs_anual['sinistros'] = (df_hs_anual['sinistros']/1000).round(1)
```

```python
fig = px.line(df_hs_anual, x='ano', y='sinistros',
              markers=True, text='sinistros',
              #height=600, width=800, #altura x largura
              line_shape="spline",
              template="plotly_dark",
              render_mode="svg",
              title = "Sinistros por Ano",
              labels=dict(ano="Ano", sinistros="Sinistros(k)")
              )
fig.update_xaxes(type="category", title=None) # se o type for date, vai respeitar o intervalo
fig.update_traces(line_width=2,textposition='top center')
fig.show()
```
